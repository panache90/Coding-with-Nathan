### Plotly Express


```python
df_plot = pd.read_csv('https://raw.githubusercontent.com/SuperDataWorld/Python/main/Data/bikerental.csv')
```

```python
fig = px.scatter(df_plot,x='casual', y='windspeed')
fig
```

```python
df_plot['season'] = df_plot['season'].astype(str)

fig = px.scatter(
    df_plot,
    x = 'casual',
    y = 'windspeed',
    color ='season',
)

fig.update_layout(
    title = 'Casual Bike Rentals V Windspeed',
    width = 1000,
    height = 500,
    xaxis_title = 'Casual Bike Rentals',
    yaxis_title = 'Windspeed',
    template="simple_white",
)
```

```python
df_plot['yr'].value_counts()
```

```python
#ONLY RUN THIS CELL ONE TIME!!!
#df_plot['yr']=df_plot['yr'].apply(lambda x: '2011' if x==0 else '2012')
```

```python
print(df_plot.iloc[-25:-1,:])
df_plot['Month']=df_plot['yr'] + "-" + df_plot['mnth'].map(str)
```

```python
bar_data = df_plot[['Month','cnt']].groupby('Month').sum().reset_index()
```

```python
bar_data.columns = ["Month of Rental","Count of Rentals"]
bar_data.head()
```

```python
fig = px.bar(bar_data, x='Month of Rental', y='Count of Rentals')
fig.update_layout(
    bargap = 0.0075,
    title = 'Bike Rentals By Month',
    width = 1000,
    height = 500,
    xaxis_title = 'Month',
    yaxis_title = 'Bike Rentals',
    template="simple_white",
    hoverlabel=dict(
        bgcolor="white",
        font_size=12,
        font_family="Arial")
    )
```
